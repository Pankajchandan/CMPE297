### Constants in Tensorflow:

```python
import tensorflow as tf
sess=tf.Session()
```

```python
c=tf.zeros([2,2],dtype=tf.int32)
print(sess.run(c))
```

```python
one_mat = tf.ones([3,4],dtype=tf.int32)
print(sess.run(one_mat))
```

```python
const = tf.constant([1.0,2.5,3.5], shape=[2,3], dtype=tf.float32)
print(sess.run(const))
```

```python
const2 = tf.ones_like(const)
print(sess.run(const2))
```

```python
const3 = tf.fill([3,4],10)
print(sess.run(const3))
```

### Variables in Tensorflow

```python
with tf.variable_scope("test",reuse=True):
    other_variable = tf.get_variable("other_variable", dtype=tf.int32,initializer=tf.constant([23, 45]))
```

```python
sess.run(tf.global_variables_initializer())
print(sess.run(other_variable))
```

```python
weights = tf.Variable(tf.random_normal([784, 200], stddev=0.35),
                      name="weights")
biases = tf.Variable(tf.zeros([200]), name="biases")

init=tf.global_variables_initializer()

sess.run(init)
print(sess.run([weights,biases]))
```

```python
other_variable2 = tf.get_variable("other_variable2", dtype=tf.int32, initializer=tf.fill([3,4,5],5))
sess.run(tf.global_variables_initializer())
print(sess.run(other_variable2))
```

### PlaceHolders

```python
import numpy as np 
holder = tf.placeholder(tf.float32,shape=[20,20])
X = np.random.rand(20,20)
print(sess.run(holder,feed_dict={holder:X}))
```

```python
holder = tf.placeholder(tf.float32,shape=[None,2])
X = np.random.rand(20,2)
print(sess.run(holder,feed_dict={holder:X}))
```

```python
holder = tf.placeholder(tf.float32,shape=[None,1])
X = [[1],[2]]
print(sess.run(holder,feed_dict={holder:X}))
```

```python
holder = tf.placeholder(tf.float32,shape=[None,5])
X = [[1,2,3,4,5],[34,56,67,78,89]]
print(sess.run(holder,feed_dict={holder:X}))
```

### Matrix Multiplication

```python
holder = tf.placeholder(tf.float32,shape=[None,2])
holder2 = tf.placeholder(tf.float32,shape=[2,None])
X = np.random.rand(10,2)
Y = np.random.rand(2,10)
mul=tf.matmul(holder,holder2)
print(sess.run(mul,feed_dict={holder:X,holder2:Y}))
```
